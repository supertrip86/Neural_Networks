Now, can there be some duplicate images in the dataset? Let's find out:

```python
hash_dict = {}

for directory in directories:
    image_files = os.listdir(os.path.join(dataset_dir, directory))

    for image_file in image_files:
        # check if there are duplicate images in the dataset based on the MD5 hash of each raw image data
        img = Image.open(os.path.join(dataset_dir, directory, image_file))
        img_hash = hashlib.md5(img.tobytes()).hexdigest()

        if img_hash not in hash_dict:
            hash_dict[img_hash] = [(directory, image_file)]
        else:
            hash_dict[img_hash].append((directory, image_file))

duplicates_found = False

for img_hash, img_locs in hash_dict.items():
    if len(img_locs) > 1:
        if not duplicates_found:
            print('Duplicate images found:')
            duplicates_found = True
        for directory, image_file in img_locs:
            print(f'Image: {image_file} in {directory}')

if not duplicates_found:
    print('No duplicate images were found in the dataset.')
```

I am now going to add modularity to the custom VGG16 model used in part (1). More specifically, I have added three early-exits (one of which is a final exit) right after each of the last three Max Pooling layers.

```python
class EarlyExitsModel(Model):
    def __init__(self, input_shape=(224, 224, 3), num_classes=10, confidence_threshold=2.3):
        super(EarlyExitsModel, self).__init__()

        self.confidence_threshold = confidence_threshold
        self.num_classes = num_classes

        self.conv1 = Conv2D(64, (3, 3), activation='relu', padding='same')
        self.bn1 = BatchNormalization()
        self.conv2 = Conv2D(64, (3, 3), activation='relu', padding='same')
        self.bn2 = BatchNormalization()
        self.maxpool1 = MaxPooling2D(pool_size=(2, 2), strides=(2, 2))

        self.conv3 = Conv2D(128, (3, 3), activation='relu', padding='same')
        self.bn3 = BatchNormalization()
        self.conv4 = Conv2D(128, (3, 3), activation='relu', padding='same')
        self.bn4 = BatchNormalization()
        self.maxpool2 = MaxPooling2D(pool_size=(2, 2), strides=(2, 2))

        # Early exit 1
        self.exit1 = self.build_exit_block(1)

        self.conv5 = Conv2D(256, (3, 3), activation='relu', padding='same')
        self.bn5 = BatchNormalization()
        self.conv6 = Conv2D(256, (3, 3), activation='relu', padding='same')
        self.bn6 = BatchNormalization()
        self.conv7 = Conv2D(256, (3, 3), activation='relu', padding='same')
        self.bn7 = BatchNormalization()
        self.maxpool3 = MaxPooling2D(pool_size=(2, 2), strides=(2, 2))

        # Early exit 2
        self.exit2 = self.build_exit_block(2)

        self.conv8 = Conv2D(512, (3, 3), activation='relu', padding='same')
        self.bn8 = BatchNormalization()
        self.conv9 = Conv2D(512, (3, 3), activation='relu', padding='same')
        self.bn9 = BatchNormalization()
        self.conv10 = Conv2D(512, (3, 3), activation='relu', padding='same')
        self.bn10 = BatchNormalization()
        self.maxpool4 = MaxPooling2D(pool_size=(2, 2), strides=(2, 2))

        # Final exit
        self.final_exit = self.build_exit_block('final')

    def build_exit_block(self, exit_num):
        layers = tf.keras.Sequential()
        layers.add(GlobalAveragePooling2D())
        layers.add(Dense(1024, activation='relu'))
        layers.add(BatchNormalization())
        layers.add(Dense(512, activation='relu'))
        layers.add(BatchNormalization())
        layers.add(Dense(self.num_classes, activation='softmax', name=f'exit_{exit_num}'))
        return layers

    def call(self, inputs, training=False):
        if training:
            x = self.conv1(inputs)
            x = self.bn1(x, training=training)
            x = self.conv2(x)
            x = self.bn2(x, training=training)
            x = self.maxpool1(x)

            x = self.conv3(x)
            x = self.bn3(x, training=training)
            x = self.conv4(x)
            x = self.bn4(x, training=training)
            x = self.maxpool2(x)

            # Early exit 1
            exit1_out = self.exit1(x)

            x = self.conv5(x)
            x = self.bn5(x, training=training)
            x = self.conv6(x)
            x = self.bn6(x, training=training)
            x = self.conv7(x)
            x = self.bn7(x, training=training)
            x = self.maxpool3(x)

            # Early exit 2
            exit2_out = self.exit2(x)

            x = self.conv8(x)
            x = self.bn8(x, training=training)
            x = self.conv9(x)
            x = self.bn9(x, training=training)
            x = self.conv10(x)
            x = self.bn10(x, training=training)
            x = self.maxpool4(x)

            # Final exit
            final_exit_out = self.final_exit(x)

            return [exit1_out, exit2_out, final_exit_out]
        else:
            x = self.conv1(inputs)
            x = self.bn1(x, training=training)
            x = self.conv2(x)
            x = self.bn2(x, training=training)
            x = self.maxpool1(x)

            x = self.conv3(x)
            x = self.bn3(x, training=training)
            x = self.conv4(x)
            x = self.bn4(x, training=training)
            x = self.maxpool2(x)

            # Early exit 1
            exit1_out = self.exit1(x)
            entropy = -tf.reduce_sum(exit1_out * tf.math.log(exit1_out + 1e-10))

            if entropy < self.confidence_threshold:
                return exit1_out, 1
            else:
                x = self.conv5(x)
                x = self.bn5(x, training=training)
                x = self.conv6(x)
                x = self.bn6(x, training=training)
                x = self.conv7(x)
                x = self.bn7(x, training=training)
                x = self.maxpool3(x)

                # Early exit 2
                exit2_out = self.exit2(x)
                entropy = -tf.reduce_sum(exit2_out * tf.math.log(exit2_out + 1e-10))

                if entropy < self.confidence_threshold:
                    return exit2_out, 2
                else:
                    x = self.conv8(x)
                    x = self.bn8(x, training=training)
                    x = self.conv9(x)
                    x = self.bn9(x, training=training)
                    x = self.conv10(x)
                    x = self.bn10(x, training=training)
                    x = self.maxpool4(x)

                    # Final exit
                    final_exit_out = self.final_exit(x)
                    return final_exit_out, 3

# I chose a confidence_threshold value that is in between 0 (min) and 2.32 (max)
model = EarlyExitsModel(input_shape=(224, 224, 3), num_classes=10, confidence_threshold=1.15)
```

To better compare the performances of the two models, I am going to maintain the same learning rate of $0.0001$. For the early-exits model, however, I am going to implement a custom loss function, which will be the **joint loss**, as suggested in ["Lecture 8: Modular and dynamic neural networks", page 14](https://www.sscardapane.it/assets/files/nnds2022/Lecture_8_Dynamic_NNs.pdf):

$$Joint\ Loss = CE(\hat{y}, y) + \sum_e{\lambda_{e} CE(\hat{y_{e}}, y)}$$

where $\lambda_e$ are the loss weights to be assigned to each early-exit.

The second part of the work is to make sure that we can properly make inference over the early-exits model. To do so, I defined a custom *call* function in the *EarlyExitsModel* where at inference time (when training = False) we compute the entropy of the prediction returned by each early-exit: if the value of the entropy is below a certain threshold $\gamma$, we can exit and return the prediction.

In short, at each early-exit we check if the condition below is respected:

$$H[y] = -\sum_c{y_c \log(y_c)} < \gamma$$

If so, we exit and return the prediction. Given that the entropy of each prediction can vary between a maximum of $2.32$ (due to the fact that I have 10 classes in my dataset) and $0$ (the entropy is going to tend to $0$ depending on how trained is the model), I chose for $\gamma$ a value that is in between $0$ and $2.32$. More specifically, I chose $1.15$.


```python
# Training loop
epochs = 20

steps_per_epoch = (train_data.samples // train_data.batch_size) + 1
val_steps_per_epoch = (validation_data.samples // validation_data.batch_size) + 1

learning_rate=0.0001
optimizer=tf.keras.optimizers.Adam(learning_rate=learning_rate)

history = {
    "train_loss": [],
    "train_accuracy": [],
    "train_loss_exit1": [],
    "train_loss_exit2": [],
    "train_loss_final_exit": [],
    "train_accuracy_exit1": [],
    "train_accuracy_exit2": [],
    "train_accuracy_final_exit": [],
    "val_loss": [],
    "val_accuracy": [],
    "val_loss_exit1": [],
    "val_loss_exit2": [],
    "val_loss_final_exit": [],
    "val_accuracy_exit1": [],
    "val_accuracy_exit2": [],
    "val_accuracy_final_exit": []
}

# Metrics to keep track of loss and accuracy during training and validation
train_loss = tf.keras.metrics.Mean(name='train_loss')
train_accuracy = tf.keras.metrics.CategoricalAccuracy(name='train_accuracy')
val_loss = tf.keras.metrics.Mean(name='val_loss')
val_accuracy = tf.keras.metrics.CategoricalAccuracy(name='val_accuracy')

# Metrics for each exit during training
train_loss_exit1 = tf.keras.metrics.Mean(name='train_loss_exit1')
train_loss_exit2 = tf.keras.metrics.Mean(name='train_loss_exit2')
train_loss_final_exit = tf.keras.metrics.Mean(name='train_loss_final_exit')
train_accuracy_exit1 = tf.keras.metrics.CategoricalAccuracy(name='train_accuracy_exit1')
train_accuracy_exit2 = tf.keras.metrics.CategoricalAccuracy(name='train_accuracy_exit2')
train_accuracy_final_exit = tf.keras.metrics.CategoricalAccuracy(name='train_accuracy_final_exit')

# Metrics for each exit during validation
val_loss_exit1 = tf.keras.metrics.Mean(name='val_loss_exit1')
val_loss_exit2 = tf.keras.metrics.Mean(name='val_loss_exit2')
val_loss_final_exit = tf.keras.metrics.Mean(name='val_loss_final_exit')
val_accuracy_exit1 = tf.keras.metrics.CategoricalAccuracy(name='val_accuracy_exit1')
val_accuracy_exit2 = tf.keras.metrics.CategoricalAccuracy(name='val_accuracy_exit2')
val_accuracy_final_exit = tf.keras.metrics.CategoricalAccuracy(name='val_accuracy_final_exit')

# Given that late exits predictions should be more accurate (in theory), I chose to weigh each loss as follows
loss_weights = [0.35, 0.65, 1]

@tf.function
def train_step(x, y):
    with tf.GradientTape() as tape:
        logits = model(x, training=True)
        loss_exit1 = tf.keras.losses.categorical_crossentropy(y, logits[0])
        loss_exit2 = tf.keras.losses.categorical_crossentropy(y, logits[1])
        loss_final_exit = tf.keras.losses.categorical_crossentropy(y, logits[2])

        # Compute the weighted loss for each exit and accumulate them (as indicated in "Lecture 8: Modular and dynamic neural networks", page 14)
        joint_loss = tf.reduce_sum([loss_weights[0] * loss_exit1, loss_weights[1] * loss_exit2, loss_weights[2] * loss_final_exit])

    grads = tape.gradient(joint_loss, model.trainable_weights)
    optimizer.apply_gradients(zip(grads, model.trainable_weights))

    return logits, joint_loss, loss_exit1, loss_exit2, loss_final_exit

@tf.function
def test_step(x, y):
    val_logits, exit_num = model(x, training=False)
    val_loss = tf.keras.losses.categorical_crossentropy(y, val_logits)

    return val_logits, exit_num, val_loss

# Training loop
for epoch in range(epochs):
    print(f'Epoch {epoch+1}/{epochs}')

    # Reset the metrics at the start of the next epoch
    train_loss.reset_states()
    train_accuracy.reset_states()
    val_loss.reset_states()
    val_accuracy.reset_states()

    train_loss_exit1.reset_states()
    train_loss_exit2.reset_states()
    train_loss_final_exit.reset_states()
    train_accuracy_exit1.reset_states()
    train_accuracy_exit2.reset_states()
    train_accuracy_final_exit.reset_states()

    val_loss_exit1.reset_states()
    val_loss_exit2.reset_states()
    val_loss_final_exit.reset_states()
    val_accuracy_exit1.reset_states()
    val_accuracy_exit2.reset_states()
    val_accuracy_final_exit.reset_states()

    # Initialize progress bar
    progbar = tf.keras.utils.Progbar(steps_per_epoch)

    # Training step
    for step, (x_batch_train, y_batch_train) in enumerate(train_data):
        # Break the loop at end of epoch
        if step >= steps_per_epoch:
            break

        logits, joint_loss, loss_exit1, loss_exit2, loss_final_exit = train_step(x_batch_train, y_batch_train)

        # Update training metrics
        train_loss.update_state(joint_loss)
        train_accuracy.update_state(y_batch_train, logits[2])

        # Update individual output metrics
        train_loss_exit1.update_state(loss_exit1)
        train_loss_exit2.update_state(loss_exit2)
        train_loss_final_exit.update_state(loss_final_exit)
        train_accuracy_exit1.update_state(y_batch_train, logits[0])
        train_accuracy_exit2.update_state(y_batch_train, logits[1])
        train_accuracy_final_exit.update_state(y_batch_train, logits[2])

        progbar.update(step, values=[("loss", train_loss.result()), ("accuracy", train_accuracy.result())])

    # Validation step
    for val_step, (x_batch_val, y_batch_val) in enumerate(validation_data):
        # Break the loop at end of epoch
        if val_step >= val_steps_per_epoch:
            break

        val_logits, exit_num, val_loss_val = test_step(x_batch_val, y_batch_val)

        # Update validation metrics
        val_loss.update_state(val_loss_val)
        val_accuracy.update_state(y_batch_val, val_logits)

        # Update individual output validation metrics
        if exit_num == 1:
            val_loss_exit1.update_state(val_loss_val)
            val_accuracy_exit1.update_state(y_batch_val, val_logits)
        elif exit_num == 2:
            val_loss_exit2.update_state(val_loss_val)
            val_accuracy_exit2.update_state(y_batch_val, val_logits)
        elif exit_num == 3:
            val_loss_final_exit.update_state(val_loss_val)
            val_accuracy_final_exit.update_state(y_batch_val, val_logits)

    # Update the progress bar one last time at the end of the epoch
    progbar.update(steps_per_epoch, values=[("val_loss", val_loss.result()), ("val_accuracy", val_accuracy.result())], finalize=True)

    for key in history.keys():
        history[key].append(locals()[key].result().numpy())
```
